#### 一幅图形中有两个子图的例子
subplot()函数不仅可以将图形分为不同的绘图区域，还能激活特定子图，以便用命令控制它

subplot()函数用参数设置分区模式和当前子图。只有当前子图会受到命令的影响。

subplot()函数的参数由三个整数组成：第1个数字决定图形沿着垂直方向被分为几个部分；第2个数字决定图形沿着水平方向被分为几个部分；第3个数字设定可以直接用命令控制的子图

```python
t=np.arange(0,5,0.1)
y1=np.sin(2*np.pi*t)
y2=np.sin(2*np.pi*t)
plt.subplot(211)
plt.plot(t,y1,'b-.')
plt.subplot(212)
plt.plot(t,y2,'r--')
plt.show()
```

## 五、处理日期值

### 使用时间尺度来管理日期。
首先需要导入matplotlib.dates模块，该模块专门用于管理日期类型的数据。

其次：定义时间尺度,利用monthlocator()和daylocator()函数，分别表示月份和日子。
将只显示年月的格式参数传给dataformatter()函数。

在xaxis对象上调用set_major_locator()函数和set_minor_locator()函数，为x轴设置两种不同的标签。

```python
import datetime
import matplotlib.dates as mdates
months=mdates.MonthLocator()#表示月份
days=mdates.DayLocator()
timeFmt=mdates.DateFormatter('%Y-%m')
events=[datetime.date(2015,1,23),datetime.date(2015,1,28),datetime.date(2015,2,23),
        datetime.date(2015,2,28),datetime.date(2015,3,23),datetime.date(2015,3,28),datetime.date(2015,4,23)]
readings=[12,22,25,21,18,13,11]
fig,ax=plt.subplots()
plt.plot(events,readings)
ax.xaxis.set_major_locator(months) # x轴只显示月份的刻度标签，可读性性更强
ax.xaxis.set_major_formatter(timeFmt)
```

## 六、图表类型

### 1、线形图
线形图：一对对(x,y)值组成的数据在图表中的位置取决于两条轴(x和y)的刻度范围

首先：创建包含x值得数组，用作x轴；然后，用np.sin()函数直接求得这一列x值所对应的y值

```python
# 画图像y=sin（n*x）/x
x=np.arange(-2*np.pi,2*np.pi,0.01)  # x轴的数据范围为-2π到2π
y=np.sin(3*x)/x
y2=np.sin(2*x)/x
y3=np.sin(5*x)/x
plt.plot(x,y,'k--')
plt.plot(x,y2,'m-')
plt.plot(x,y3,color='#87a3cc')
# 颜色编码;b蓝色，g绿色，r红色，c蓝绿色，m洋红，y黄色，k黑色，w白色
plt.show()
```

### 3、条状图bar()

```python
index=[0,1,2,3,4]
values=[4,7,8,9,13]
std1=[0.3,0.4,0.1,0.2,0.6]
plt.bar(index,values,yerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='First')
# 将包含标准差的列表传给yerr关键字参数，就能添加标准差，
# error_kw用于显示误差线的关键字参数；ecolor：指定误差线的颜色，
# capsize指定误差线两头横线的宽度
plt.xticks(index,['a','b','c','d','e'])# xticks表示刻度位置
plt.title('A bar chart')
plt.legend(loc=2)
plt.show()
```

带有误差线的条状图

#### 3.1、水平条状图 barh()
类别分布在y轴上，数值显示在x轴上

```python
index=[0,1,2,3,4]
values=[4,7,8,9,13]
std1=[0.3,0.4,0.1,0.2,0.6]
plt.title('A bar chart')
plt.barh(index,values,xerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='Second')
plt.yticks(index,['A','B','C','D','E'])
plt.legend(loc=4)
plt.show()
# 注意水平条状图将误差赋值给xerr，条状图赋值给yerr
```

#### 3.2 多序列条状图
首先定义一列索引值，把它们分配给x轴，每个值对应一条状图形，索引值代表类别。（多序列条状图：多个长条共用相同的类别）

解决方法：把每个类别占据的空间分为多个部分。想要显示几个长条，就将其分为几部分

```python
import matplotlib.pyplot as plt
import numpy as np
index=np.arange(5)
values1=[4,3,2,6,5]
values2=[6,6,7,4,8]
values3=[1,2,4,3,1]
bw=0.3
plt.axis([0,5,0,9])
plt.title('A multiseries bar chart',fontsize=20)
plt.bar(index,values1,bw,color='r')
plt.bar(index+bw,values2,bw,color='g')
plt.bar(index+2*bw,values3,bw,color='b')
plt.xticks(index+1*bw,['a','b','c','d','e'])
plt.show()
```

#### 3.3 多序列堆积条状图
将几个条状图形堆积在一起形成一个更大的长条。如果想要表示总和是由几个条状图相加得到的，堆积图特别合适。

方法：在每个bar()函数中添加bottom关键字参数，把每个序列赋给相应的bottom关键字参数。

```python
import numpy as np
series1=np.array([2,3,4,5])
series2=np.array([1,2,3,4])
series3=np.array([1,4,2,1])
index=np.arange(4)
plt.axis([0,4,0,15])
plt.bar(index+0.5,series1,color='g')
plt.bar(index+0.5,series2,color='b',bottom=series1)
plt.bar(index+0.5,series3,color='r',bottom=(series1+series2))
plt.xticks(index+0.5,['Jan','Feb','Mar','Apr'])
# bar中的参数index是指条状图的位置，
# xticks中的参数index是指x轴标签的位置
# bar和xticks的起始位置是从y轴（0点）开始
plt.show()
```

```python
series1=np.array([2,3,4,5])
series2=np.array([1,2,3,4])
series3=np.array([1,4,2,1])
index=np.arange(4)
plt.axis([0,14,0,4])
plt.barh(index+0.5,series1,color='g')
plt.barh(index+0.5,series2,color='b',left=series1)
plt.barh(index+0.5,series3,color='r',left=(series1+series2))
# 水平条状图用left堆叠，竖直条状图用bottom堆叠
plt.yticks(index+0.5,['Jan','Feb','Mar','Apr'])
plt.show()
```

```python
# 不同的影线填充条状图
# 方法：首先将条状图颜色设置为白色，然后用hatch关键字参数指定影线的类型。
series1=np.array([2,3,4,5])
series2=np.array([1,2,3,4])
series3=np.array([1,4,2,1])
index=np.arange(4)
plt.axis([0,4,0,15])
plt.bar(index+0.5,series1,color='w',hatch='+')
plt.bar(index+0.5,series2,color='w',hatch='//',bottom=series1)
plt.bar(index+0.5,series3,color='w',hatch='\\\\',bottom=(series1+series2))
# 同一符号出现的次数越多，则形成的阴影的线条越密集
plt.xticks(index+0.5,['Jan','Feb','Mar','Apr'])
plt.show()
```

### 4、饼图pie()
通过关键字参数设置关键特征：

axis()函数用字符串‘equal’作为参数：绘制标准的圆形饼图

colors：赋予颜色；

labels：为饼图的每一小块添加标签；

startangle：调整饼图的旋转角度（该参数接收0-360的整数）。0，表示从0开始逆时针转，为第一块。一般选择从90度开始比较好看

explode：抽取某一块饼图
（1表示这一块完全脱离饼图，0：没有抽取，0-1：表示这一块未完全脱离）

autopct：在每一块中间位置添加文本标签来显示百分比。文本格式，%3.1f%%表示小数有三位，整数有一位的浮点数

```python
labels=['Nokis','Samsung','Apple','Lumia']
values=[10,30,45,15]
colors=['y','r','g','b']
explode=[0.3,0,0,0]
plt.title('A Pie Chart')
plt.pie(values,labels=labels,colors=colors,explode=explode,shadow=True,autopct='%1.1f%%',startangle=180)
plt.axis('equal')
plt.show()
```

##### 为DataFrame绘制饼图

### 1、3D曲面

```python
fig=plt.figure()
ax=Axes3D(fig)
x=np.arange(-2,2,0.1)
y=np.arange(-2,2,0.1)
x,y=np.meshgrid(x,y)
def f(x,y):
    return (1-y**5+x**5)*np.exp(-x**2-y**2)
ax.plot_surface(x,y,f(x,y),rstride=1,cstride=1,cmap=plt.cm.hot)
#plot_surface():绘制曲面；cmap：指定各颜色；plt.cm.hot:颜色表
ax.view_init(elev=30,azim=125)
# view_init（）旋转曲面，elev：指定从哪个高度查看曲面；azim：指定曲面旋转的角度
plt.show()
```

### 八、多面板图形

#### 1、在其他子图中显示子图
把图表放入框架，在其他图表中显示。

把主Axes对象（也就是主图表）跟放置另一个Axes对象实例的框架分开。用figure()函数取到figure对象，用add_axes()函数在它上面定义两个Axes对象

```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
inner_ax=fig.add_axes([0.6,0.6,0.25,0.25])
x1=np.arange(10)
y1=np.array([1,2,7,3,4,4,3,2,1,7])
x2=np.arange(10)
y2=np.array([1,2,2,2,3,4,2,1,1,1])
ax.plot(x1,y1)
inner_ax.plot(x2,y2)
plt.show()
```

#### 2、子图网格
要把图形分成多个区域，添加多个子图，可以用subplots()函数；matplotlib中的GridSpec()h函数可以用来管理更为复杂的情况

```python
gs=plt.GridSpec(3,3)# 分为3行3列
fig=plt.figure(figsize=(6,6))
x1=np.array([1,3,2,5])
y1=np.array([3,5,3,6])
x2=np.arange(5)
y2=np.array([3,2,5,6,6])
s1=fig.add_subplot(gs[1,:2])# 用add_subplot()函数返回axes对象，s1放在第2行，占第1和第2列
s1.plot(x1,y1,'r')  # 在返回的axes对象上调用plot函数,绘制相应的图形
s2=fig.add_subplot(gs[0,:2])# s2放在第1行，占第1和第2列
s2.bar(x2,y2)
s3=fig.add_subplot(gs[2,0])#s3放在第3行，占第1列
s3.barh(x2,y2,color='g')
s4=fig.add_subplot(gs[:2,2])#s4放在第3列，占第1和第2行
s4.plot(x2,y2,'k')
s5=fig.add_subplot(gs[2,1:])# s5放在第3行，占第2和第3列
s5.plot(x1,y1,'b',x2,y2,'yo')
plt.show()
```
